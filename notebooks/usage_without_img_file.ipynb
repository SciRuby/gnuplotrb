```ruby
require 'pilot-gnuplot'
include Gnuplot

# To use gnuplot this way iRuby should be modified 
# (see https://github.com/dilcom/pilot-gnuplot/tree/master/notebooks#advanced-not-recommended).
cartesian = Plot.new(
  ['x*sin(x)', with: 'lines', lt: { rgb: 'blue', lw: 3 }],
  xrange: -10..10,
  title: 'Math function example', 
  ylabel: 'x',
  xlabel: 'x*sin(x)'
)
```

```ruby
# you can also set some term options here:
bigger_plot = cartesian.term('svg', size: [800, 800])
bigger_plot
```
