```ruby
require 'pilot-gnuplot'
include Gnuplot
x = 30.times.map { |xx| xx/3.0 }
data = [x, x.map { |xx| Math.sin(xx)*xx**2 }]
plot = Plot.new([data, with: 'lines', title: 'Some data'], title: 'Wow')
```

```ruby
x1 = (31..60).map { |xx| xx/3.0 }
data1 = [x1, x1.map { |xx| Math.sin(xx)*xx**2 }]
updated_plot = plot.update_dataset(data: data1, title: 'Updated data', with: 'points')
```
